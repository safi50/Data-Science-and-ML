#Importing Iris Dataset and Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

##Loading Data
###-- 4 Features --

1,2 : length and width of Petals

3,4 : length and width of Sepals

```python
iris = load_iris()

X,y = iris.data, iris.target

X.shape, y.shape
```

## Classification

```python
clf = DecisionTreeClassifier()

clf.fit(X,y)

clf.predict([[5,3,5,5]])
```

## Visualization

```python
plot_tree(clf)
```

## Using Matplotlib

```python
plt.figure(figsize=(10,10))
plot_tree(clf, filled = True,
          feature_names = ["Sepal Len", "Sepal Width", "Petal Len", "Petal Width" ],
          class_names = ["Iris Setosa", "Iris Versicolor", "Iris Virginica"]
          )

plt.show()
```
